```python
from collections import deque

T = int(input())
for _ in range(T):
    N = int(input())
    number_list = list(map(int, input().split()))

    number = {}
    for x in number_list:
        number[x] = number.get(x, 0) + 1

    break_list = {}

    for num in number:
        break_list[num] = {}
        queue = deque()
        queue.append((num, 0))
        visited = set()
        visited.add(num)

        while queue:
            now, count = queue.popleft()

            if now not in break_list[num]:
                break_list[num][now] = count
            else:
                if count < break_list[num][now]:
                    break_list[num][now] = count
                else:
                    continue  # 이미 더 작은 count로 방문했으면 skip

            if now == 1:
                continue
            if now % 2 == 1:
                next1 = now // 2
                next2 = next1 + 1
                if next1 not in visited:
                    queue.append((next1, count + 1))
                    visited.add(next1)
                if next2 not in visited:
                    queue.append((next2, count + 1))
                    visited.add(next2)
            else:
                next1 = now // 2
                if next1 not in visited:
                    queue.append((next1, count + 1))
                    visited.add(next1)

    min_total = float('inf')
    for candidate in break_list[number_list[0]]:
        flag = 0
        total = 0
        for num in number:
            if candidate in break_list[num]:
                flag += 1
                total += break_list[num][candidate] * number[num]
            else:
                break
        if flag == len(number):
            min_total = min(min_total, total)
    print(min_total)
```
